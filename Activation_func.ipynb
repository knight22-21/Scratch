```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def leaky_relu(z, alpha=0.01):
    return np.where(z > 0, z, alpha * z)

def prelu(z, a=0.25):  # a is a learnable parameter, hard-coded here
    return np.where(z > 0, z, a * z)

def elu(z, alpha=1.0):
    return np.where(z > 0, z, alpha * (np.exp(z) - 1))

def swish(z):
    return z * sigmoid(z)
```

```python
z = np.linspace(-10, 10, 1000)

plt.figure(figsize=(12, 10))

# ReLU
plt.subplot(3, 3, 1)
plt.plot(z, relu(z))
plt.title("ReLU")
plt.grid()

# Sigmoid
plt.subplot(3, 3, 2)
plt.plot(z, sigmoid(z))
plt.title("Sigmoid")
plt.grid()

# Tanh
plt.subplot(3, 3, 3)
plt.plot(z, tanh(z))
plt.title("Tanh")
plt.grid()

# Leaky ReLU
plt.subplot(3, 3, 4)
plt.plot(z, leaky_relu(z))
plt.title("Leaky ReLU (=0.01)")
plt.grid()

# PReLU
plt.subplot(3, 3, 5)
plt.plot(z, prelu(z))
plt.title("Parametric ReLU (a=0.25)")
plt.grid()

# ELU
plt.subplot(3, 3, 6)
plt.plot(z, elu(z))
plt.title("ELU (=1.0)")
plt.grid()

# Swish
plt.subplot(3, 3, 7)
plt.plot(z, swish(z))
plt.title("Swish")
plt.grid()

plt.tight_layout()
plt.show()
```
