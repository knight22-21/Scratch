```python
import numpy as np
```

```python
class NN:
    def __init__(self,input_size,hidden_size,output_size):
        self.w1 = np.random.randn(input_size,hidden_size) * 0.01
        self.b1 = np.zeros((1,hidden_size))
        self.w2 = np.random.randn(hidden_size,output_size) * 0.01
        self.b2 = np.zeros((1,output_size))

    def sigmoid(self,z):
        return 1/(1+ np.exp(-z))

    def sig_der(self,z):
        return z*(1-z)

    def relu(self,z):
        return np.maximum(0,z)

    def relu_der(self,z):
        return (z>0).astype(float)

    def forward(self,X):
        self.Z1 = X.dot(self.w1) + self.b1
        self.A1 = self.relu(self.Z1)

        self.Z2 = self.A1.dot(self.w2) + self.b2
        self.A2 = self.sigmoid(self.Z2)

        return self.A2

    def compute_loss(self, y_true, y_pred):
        m = y_true.shape[0]
        return -np.mean(y_true * np.log(y_pred + 1e-8) + (1-y_true) * np.log(1-y_true + 1e-8))

    def backward(self,X,y_true, y_pred, lr):
        m = y_true.shape[0]
        # Output layer gradients
        dz2 = y_pred - y_true
        dw2 = self.A1.T.dot(dz2) / m
        db2 = np.sum(dz2,axis=0, keepdims=True) / m
        # Hidden layer gradients
        da1= dz2.dot(self.w2.T)
        dz1 = da1*self.relu_der(self.Z1)
        dw1 = X.T.dot(dz1) / m
        db1 = np.sum(dz1,axis=0, keepdims=True) / m

        self.w2 -= lr*dw2
        self.b2 -= lr*db2
        self.w1 -= lr*dw1
        self.b1 -= lr*db1

    def train(self,X,y, epochs=1000, lr=0.1, verbose = True):
        for i in range (1000):
            y_preds = self.forward(X)
            loss = self.compute_loss(y,y_preds)
            self.backward(X,y,y_preds,lr)
            if verbose and i % 100 == 0:
                print(f"Epoch {i}, Loss: {loss:.4f}")


    def predict(self,x):
        y_pred = self.forward(x)
        return (y_pred > 0.5).astype(int)
            
```

```python
 if __name__ == "__main__":
    # Simple synthetic binary classification
    from sklearn.datasets import make_moons
    import matplotlib.pyplot as plt

    X, y = make_moons(n_samples=500, noise=0.2, random_state=42)
    y = y.reshape(-1, 1)

    model = NN(input_size=2, hidden_size=5, output_size=1)
    model.train(X, y, epochs=1000, lr=0.1)

    # Plot decision boundary
    h = 0.01
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap='Paired', alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y.ravel(), cmap='Paired')
    plt.title("ANN Decision Boundary")
    plt.show()
```
