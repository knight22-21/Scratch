```python
import numpy as np
import matplotlib.pyplot as plt

# === ACTIVATIONS ===
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_deriv(x):
    return 1 - np.tanh(x) ** 2
```

```python
class DenseLayer:
    def __init__(self, in_dim, out_dim, activation='tanh'):
        self.W = np.random.randn(in_dim, out_dim) * 0.01
        self.b = np.zeros((1, out_dim))
        self.activation = activation

    def forward(self, x):
        self.input = x
        self.z = x @ self.W + self.b
        if self.activation == 'tanh':
            self.out = tanh(self.z)
        elif self.activation == 'sigmoid':
            self.out = sigmoid(self.z)
        return self.out

    def backward(self, grad_output):
        if self.activation == 'tanh':
            grad = grad_output * tanh_deriv(self.z)
        elif self.activation == 'sigmoid':
            grad = grad_output * sigmoid_deriv(self.z)

        self.dW = self.input.T @ grad
        self.db = np.sum(grad, axis=0, keepdims=True)
        return grad @ self.W.T  # Return gradient for previous layer

    def step(self, lr):
        self.W -= lr * self.dW
        self.b -= lr * self.db
```

```python
class Generator:
    def __init__(self, input_dim=2, hidden_dim=4, output_dim=2):
        self.fc1 = DenseLayer(input_dim, hidden_dim, activation='tanh')
        self.fc2 = DenseLayer(hidden_dim, output_dim, activation='tanh')

    def forward(self, z):
        h = self.fc1.forward(z)
        out = self.fc2.forward(h)
        return out

    def backward(self, grad, lr):
        grad = self.fc2.backward(grad)
        grad = self.fc1.backward(grad)
        self.fc1.step(lr)
        self.fc2.step(lr)
```

```python
class Discriminator:
    def __init__(self, input_dim=2, hidden_dim=4):
        self.fc1 = DenseLayer(input_dim, hidden_dim, activation='tanh')
        self.fc2 = DenseLayer(hidden_dim, 1, activation='sigmoid')

    def forward(self, x):
        h = self.fc1.forward(x)
        out = self.fc2.forward(h)
        return out

    def backward(self, grad, lr):
        grad = self.fc2.backward(grad)
        grad = self.fc1.backward(grad)
        self.fc1.step(lr)
        self.fc2.step(lr)
```

```python
class GAN:
    def __init__(self, lr=0.01):
        self.G = Generator()
        self.D = Discriminator()
        self.lr = lr

    def sample_real(self, n):
        return np.random.normal(loc=4.0, scale=1.5, size=(n, 2))

    def sample_noise(self, n):
        return np.random.randn(n, 2)

    def bce_loss(self, y_pred, y_true):
        eps = 1e-10
        return -np.mean(y_true * np.log(y_pred + eps) + (1 - y_true) * np.log(1 - y_pred + eps))

    def train(self, epochs=10000, batch_size=32, print_every=1000):
        for epoch in range(epochs):
            # === Train Discriminator ===
            real_data = self.sample_real(batch_size)
            z = self.sample_noise(batch_size)
            fake_data = self.G.forward(z)

            D_real = self.D.forward(real_data)
            D_fake = self.D.forward(fake_data)

            y_real = np.ones((batch_size, 1))
            y_fake = np.zeros((batch_size, 1))

            loss_D = self.bce_loss(D_real, y_real) + self.bce_loss(D_fake, y_fake)

            d_real_grad = D_real - y_real
            d_fake_grad = D_fake - y_fake

            self.D.backward(d_real_grad, self.lr)
            self.D.backward(d_fake_grad, self.lr)

            # === Train Generator ===
            z = self.sample_noise(batch_size)
            fake_data = self.G.forward(z)
            D_fake = self.D.forward(fake_data)

            loss_G = self.bce_loss(D_fake, y_real)  # Want D_fake close to 1

            g_grad = D_fake - y_real
            grad_through_D = self.D.fc2.backward(g_grad)
            grad_through_D = self.D.fc1.backward(grad_through_D)
            self.G.backward(grad_through_D, self.lr)

            # Update Discriminator again
            self.D.fc2.step(self.lr)
            self.D.fc1.step(self.lr)

            if epoch % print_every == 0:
                print(f"Epoch {epoch} | D_loss: {loss_D:.4f} | G_loss: {loss_G:.4f}")

    def plot(self):
        z = self.sample_noise(1000)
        fake_data = self.G.forward(z)
        real_data = self.sample_real(1000)

        plt.figure(figsize=(6, 6))
        plt.scatter(real_data[:, 0], real_data[:, 1], alpha=0.5, label='Real')
        plt.scatter(fake_data[:, 0], fake_data[:, 1], alpha=0.5, label='Fake')
        plt.legend()
        plt.grid(True)
        plt.title("Generated vs Real Data")
        plt.show()
```

```python
gan = GAN(lr=0.01)
gan.train(epochs=10000, batch_size=32, print_every=1000)
gan.plot()
```
