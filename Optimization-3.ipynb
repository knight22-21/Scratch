```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data: y = 2x + 3 + noise
np.random.seed(42)
X = np.linspace(0, 1, 100).reshape(-1, 1)
y = 2 * X + 3 + np.random.randn(*X.shape) * 0.1
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add bias term
true_line = 2 * X + 3

# Loss and gradients
def compute_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def compute_gradients(X, y, weights):
    preds = X.dot(weights)
    errors = preds - y
    grads = 2 / X.shape[0] * X.T.dot(errors)
    return grads

# --- Optimizers ---
class RMSProp:
    def __init__(self, lr=0.01, gamma=0.9, epsilon=1e-8):
        self.lr = lr
        self.gamma = gamma
        self.epsilon = epsilon
        self.E_g2 = None

    def step(self, weights, grads):
        if self.E_g2 is None:
            self.E_g2 = np.zeros_like(grads)
        self.E_g2 = self.gamma * self.E_g2 + (1 - self.gamma) * grads**2
        weights -= self.lr * grads / (np.sqrt(self.E_g2) + self.epsilon)
        return weights

class Adam:
    def __init__(self, lr=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.m = None
        self.v = None
        self.t = 0

    def step(self, weights, grads):
        if self.m is None:
            self.m = np.zeros_like(grads)
            self.v = np.zeros_like(grads)
        self.t += 1
        self.m = self.beta1 * self.m + (1 - self.beta1) * grads
        self.v = self.beta2 * self.v + (1 - self.beta2) * (grads ** 2)
        m_hat = self.m / (1 - self.beta1 ** self.t)
        v_hat = self.v / (1 - self.beta2 ** self.t)
        weights -= self.lr * m_hat / (np.sqrt(v_hat) + self.epsilon)
        return weights

class AdamW:
    def __init__(self, lr=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8, weight_decay=0.01):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.weight_decay = weight_decay
        self.m = None
        self.v = None
        self.t = 0

    def step(self, weights, grads):
        if self.m is None:
            self.m = np.zeros_like(grads)
            self.v = np.zeros_like(grads)
        self.t += 1
        self.m = self.beta1 * self.m + (1 - self.beta1) * grads
        self.v = self.beta2 * self.v + (1 - self.beta2) * (grads ** 2)
        m_hat = self.m / (1 - self.beta1 ** self.t)
        v_hat = self.v / (1 - self.beta2 ** self.t)
        weights -= self.lr * self.weight_decay * weights
        weights -= self.lr * m_hat / (np.sqrt(v_hat) + self.epsilon)
        return weights

class Nadam:
    def __init__(self, lr=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.m = None
        self.v = None
        self.t = 0

    def step(self, weights, grads):
        if self.m is None:
            self.m = np.zeros_like(grads)
            self.v = np.zeros_like(grads)
        self.t += 1
        self.m = self.beta1 * self.m + (1 - self.beta1) * grads
        self.v = self.beta2 * self.v + (1 - self.beta2) * (grads ** 2)
        m_hat = self.m / (1 - self.beta1 ** self.t)
        v_hat = self.v / (1 - self.beta2 ** self.t)
        nesterov_m = self.beta1 * m_hat + ((1 - self.beta1) * grads) / (1 - self.beta1 ** self.t)
        weights -= self.lr * nesterov_m / (np.sqrt(v_hat) + self.epsilon)
        return weights

# --- Training loop for each optimizer ---
optimizers = {
    'RMSProp': RMSProp(lr=0.01),
    'Adam': Adam(lr=0.01),
    'AdamW': AdamW(lr=0.01),
    'Nadam': Nadam(lr=0.01)
}

loss_histories = {}
fit_lines = {}

for name, optimizer in optimizers.items():
    weights = np.random.randn(2, 1)
    losses = []
    lines = []
    for epoch in range(100):
        preds = X_b.dot(weights)
        loss = compute_loss(y, preds)
        losses.append(loss)
        grads = compute_gradients(X_b, y, weights)
        weights = optimizer.step(weights, grads)

        if epoch % 20 == 0 or epoch == 99:
            lines.append(X_b.dot(weights))
    loss_histories[name] = losses
    fit_lines[name] = lines

# --- Plotting ---
fig, axes = plt.subplots(2, 4, figsize=(18, 8))

# Loss curves
for i, name in enumerate(optimizers.keys()):
    axes[0, i].plot(loss_histories[name], label='Loss')
    axes[0, i].set_title(f"{name} Loss")
    axes[0, i].set_xlabel("Epoch")
    axes[0, i].set_ylabel("Loss")
    axes[0, i].legend()

# Fit lines
for i, name in enumerate(optimizers.keys()):
    axes[1, i].scatter(X, y, label="Data", s=10)
    axes[1, i].plot(X, true_line, 'g--', label="True Line")
    for j, line in enumerate(fit_lines[name]):
        alpha = 0.3 + 0.7 * (j / len(fit_lines[name]))
        axes[1, i].plot(X, line, color="blue", alpha=alpha)
    axes[1, i].set_title(f"{name} Fit Progress")
    axes[1, i].legend()

plt.tight_layout()
plt.show()
```
