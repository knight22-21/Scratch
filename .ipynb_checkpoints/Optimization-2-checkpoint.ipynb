```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to minimize
def f(x):
    return x**2

# Gradient of f(x)
def grad_f(x):
    return 2 * x
```

```python
def gradient_descent(x0, lr=0.1, steps=50):
    x = x0
    trajectory = [x]

    for _ in range(steps):
        grad = grad_f(x)
        x = x - lr * grad
        trajectory.append(x)

    return trajectory
```

```python
def momentum_optimizer(x0, lr=0.1, gamma=0.9, steps=50):
    x = x0
    v = 0
    trajectory = [x]

    for _ in range(steps):
        grad = grad_f(x)
        v = gamma * v + lr * grad
        x = x - v
        trajectory.append(x)

    return trajectory
```

```python
def nesterov_optimizer(x0, lr=0.1, gamma=0.9, steps=50):
    x = x0
    v = 0
    trajectory = [x]

    for _ in range(steps):
        lookahead = x - gamma * v
        grad = grad_f(lookahead)
        v = gamma * v + lr * grad
        x = x - v
        trajectory.append(x)

    return trajectory
```

```python
def adagrad_optimizer(x0, lr=1.0, eps=1e-8, steps=50):
    x = x0
    G = 0
    trajectory = [x]

    for _ in range(steps):
        grad = grad_f(x)
        G += grad ** 2
        adjusted_lr = lr / (np.sqrt(G) + eps)
        x = x - adjusted_lr * grad
        trajectory.append(x)

    return trajectory
```

```python
def plot_trajectories(trajectories, labels):
    for trajectory, label in zip(trajectories, labels):
        plt.plot(trajectory, label=label)

    plt.axhline(y=0, color='k', linestyle='--')
    plt.xlabel('Iterations')
    plt.ylabel('x')
    plt.legend()
    plt.title('Optimizer Trajectories')
    plt.grid(True)
    plt.show()

# Initialize
x0 = 10

# Run all optimizers
gd = gradient_descent(x0)
momentum = momentum_optimizer(x0)
nag = nesterov_optimizer(x0)
adagrad = adagrad_optimizer(x0)

# Plot them
plot_trajectories(
    [gd, momentum, nag, adagrad],
    ['Gradient Descent', 'Momentum', 'NAG', 'Adagrad']
)
```
